```python
from matplotlib import pyplot as plt
from scipy.stats import uniform
import numpy as np
```

```python
a = 1
b = 5
x = np.linspace(a - 1, b + 1, 100)
```

```python
dist_u = uniform(loc=a, scale=abs(a - b))
```

```python
figure, ax = plt.subplots(figsize=(8, 8))
ax.grid(linewidth=0.5)
plt.title('Uniform Distribution function')
plt.plot(x, dist_u.cdf(x), color='r', label=r'F({0}, {1})'.format(a, b))
plt.legend()
```

```python
figure, ax = plt.subplots(figsize=(8, 8))
ax.grid(linewidth=0.5)
plt.title('Probability density function')
plt.plot(x, dist_u.pdf(x), color='b', label=r'f({0}, {1})'.format(a, b))
plt.legend()
```

```python
from scipy.stats import poisson
```

```python
k = 10
x = np.arange(0, 30)
```

```python
dist_p = poisson(k)
```

```python
figure, ax = plt.subplots(figsize=(8, 8))
ax.grid(linewidth=0.5)
plt.title('Poisson Distribution function')
plt.plot(x, dist_p.cdf(x), color='b', label=r'F({0})'.format(k))
plt.legend()
```

```python
figure, ax = plt.subplots(figsize=(8, 8))
ax.grid(linewidth=0.5)
plt.title('Probability density function')
plt.plot(x, dist_p.pmf(x), color='r', label=r'f({0})'.format(k))
plt.legend()
```
